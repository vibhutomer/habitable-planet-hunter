```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('full_data.csv')
#Use only 30 columns

#extract those 30 columns to df

df = data[[
    'P_MASS','P_RADIUS','P_DENSITY','P_GRAVITY','P_ESCAPE','P_TYPE',
    'P_PERIOD','P_SEMI_MAJOR_AXIS','P_ECCENTRICITY','P_INCLINATION',
    'P_OMEGA','P_PERIASTRON','P_APASTRON','P_IMPACT_PARAMETER','P_HILL_SPHERE',
    'S_MASS','S_RADIUS','S_LUMINOSITY','S_TEMPERATURE','S_AGE',
    'S_METALLICITY','S_LOG_G','S_TYPE','S_MAG','S_DISC','S_MAGNETIC_FIELD',
    'S_SNOW_LINE','S_TIDAL_LOCK','P_DETECTION','P_DISTANCE'
]]
```

I will use a for loop now to print the Distplot for every column 5 to 29 columns

```python
#a for loop for every column distplot

for cols in df.columns[5:29]:
    if df[cols].dtype != 'object':
      sns.distplot(df[cols])
      plt.ylabel(cols)
      plt.show()
    else:
      print(f"Skipping distplot for '{cols}' as it is an object column.")
```

```python
# Similarly BoxPlot for every 5:29

for cols in df.columns[5:29]:
    if df[cols].dtype != 'object':
        # Drop NaN values for the current column
        plot_data = df[cols].dropna()
        # Only plot if there's actual data after dropping NaNs
        if not plot_data.empty:
            plt.figure(figsize=(8, 6))
            sns.boxplot(y=plot_data)
            plt.ylabel(cols, fontsize=12)
            plt.title(f'Boxplot of {cols}', fontsize=14)
            plt.grid(axis='y', linestyle='--', alpha=0.7)
            plt.show()
        else:
            print(f"Skipping boxplot for '{cols}' as it contains no valid data.")
```

**Categorical Data**

```python
cat_cols = ['P_TYPE','P_DETECTION']

for col in cat_cols:
    plt.figure(figsize=(7, 4))
    sns.countplot(data=df, x=col)
    plt.title(f'Count Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
```

```python
#Observational bias check
#Closer systems and brighter stars are easier to detect.

plt.figure(figsize=(6, 5))
plt.scatter(df['P_DISTANCE'], df['S_MAG'], alpha=0.6)
plt.xlabel('Planet Distance')
plt.ylabel('Stellar Magnitude')
plt.gca().invert_yaxis()   # lower magnitude = brighter star
plt.show()
```
